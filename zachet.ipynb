## 1.1. Скачать данные по ссылке https://www.kaggle.com/datasets/ionaskel/laptop-prices



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('kc-house-data.csv')
df.sample(15)
```

### 1.2 Изучите стоимости недвижимости


```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='price', bins=20, kde=True)
plt.title('Распределение стоимостей недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Частота')
plt.show()
```

## На гистограмме видно, что большинство домов имеют стоимость до 1 миллиона долларов. Наблюдается небольшой хвост с более высокими стоимостями, что может указывать на наличие роскошных или особенных объектов недвижимости в выборке.

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='sqft_living', bins=20, kde=True)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Частота')
plt.show()
```

## Гпафик демонстрирует, что большинство домов имеют площадь в пределах 2000-3000. Относительно небольшое количество домов имеют более высокую площадь, что может указывать на более крупные или роскошные объекты недвижимости.

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='yr_built', bins=30, kde=False)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Частота')
plt.show()
```

## На гистограмме видно, что большинство домов были построены в период с 1950 по 2000 год. Это может указывать на популярность строительства в данном временном диапазоне.

### 2.2 Изучите распределение этажей домов


```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='floors')
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество')
plt.show()
```

## Большинство домов имеют один или два этажа. Домов с тремя этажами и более намного меньше, что может указывать на более редкие и возможно более роскошные объекты недвижимости.

### 2.3 Изучите распределение состояния домов
